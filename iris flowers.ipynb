## **SPLITTING DATA**

```python
from sklearn.model_selection import train_test_split
```

```python
X= dataset.drop(columns=['class'])
Y= dataset['class']

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.30)
```

## LOGISTIC REGRESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
prediction1 = model1.predict(x_test)
```

**ACCURACY**

```python
print("ACCURACY: ", model1.score(x_test,y_test)*100)
```

```python
from sklearn.metrics import accuracy_score
print(f'LogisiticRegression Accuracy: {accuracy_score(y_test, prediction1)}')
```

**CONFUSION MATRIX**

```python
from sklearn.metrics import confusion_matrix
print(f"""
 Confusion Matrix: 
{confusion_matrix(y_test, prediction1)}
""")
```

**CLASSIFICATION REPORT**

```python
from sklearn.metrics import classification_report
print(f'Classification Report: \n {classification_report(y_test, prediction1)}')
```

```python
comparison = pd.DataFrame({'Actual Values':y_test,'Predicted Values':prediction1})
comparison
```

## DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
```

```python
model2.fit(x_train,y_train)
```

### VISUALISING DECISION TREE

```python
from sklearn.tree import plot_tree, export_text
```

```python
plt.figure(figsize=(30,15))

# Visualize the tree graphically using plot_tree
plot_tree(model2, feature_names=x_train.columns, max_depth=3,filled=True);
```

```python
tree_text = export_text(model2, max_depth=10, feature_names=list(x_train.columns))
```

```python
print(tree_text[:2000])
```

### IMPORTANT FEATURES 

```python
tree_importances = model2.feature_importances_
```

```python
tree_importance_df = pd.DataFrame({
    'feature': x_train.columns,
    'importance': tree_importances
}).sort_values('importance', ascending=False)
```

```python
plt.title('Decision Tree Feature Importance')
sns.barplot(data=tree_importance_df.head(10), x='importance', y='feature');
```
